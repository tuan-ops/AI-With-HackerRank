```python
import numpy as np
x = [15, 12, 8, 8, 7, 7, 7, 6, 5, 3]
y = [10, 15, 17, 11, 13, 17, 20, 13, 9, 15]
x_m = np.mean(x)
y_m = np.mean(y)
tu_so = np.sum((x-x_m) * (y-y_m))
mau_so = np.sqrt(np.sum((x-x_m)**2) * np.sum((y-y_m)**2))
r = tu_so/mau_so
print(r)
```
