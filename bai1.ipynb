-   Charlie đang muốn mua một căn nhà và đã thu thập các đặc trưng trong khu vực
-   Với mỗi căn nhà sẽ ghi lại đặc trưng trên thang từ 0 đến 1 cùng với giá trên mỗi foot
-   Tuy nhiên một số căn nhà bị thiếu dữ liệu về giá, bạn cần ước lượng giá trên mỗi foot vuông dựa trên dữ liệu đặc trưng có sẵn và thông tin giá của các căn khác
-   Dữ liệu được tổ chức như sau
    +   Có F đặc trưng cho mỗi căn nhà
    +   Mỗi dòng chứa F giá trị đặc trưng sau đó là giá trên mỗi foot vuông (tổng cộng F + 1 cột)
    +   Charlie có dữ liệu của N căn nhà, tạo thành bảng gồm N dòng và (F + 1) cột

```python
import numpy as np
import pandas as pd
import sys
import statsmodels as stats
```

```python
data = """2 7
0.18 0.89 109.85
1.0 0.26 155.72
0.92 0.11 137.66
0.07 0.37 76.17
0.85 0.16 139.75
0.99 0.41 162.6
0.87 0.47 151.77
4
0.49 0.18
0.57 0.8
0.56 0.64
0.76 0.18
"""

data = list(map(float, data.split()))
i = 0
F = int(data[i]); i+=1
N = int(data[i]); i+=1
train = []
for _ in range(N):
    row = data[i: i + F + 1]
    train.append(row)
    i+= F + 1
df_train = pd.DataFrame(train)
X = df_train.iloc[:, :-1].values
y = df_train.iloc[:,-1].values
T = int(data[i]); i+=1
test = []
for _ in range(T):
    row = data[i : i + F]
    test.append(row)
    i += F
X_test = np.array(test)
X = np.hstack([np.ones((N, 1)), X])
X_test = np.hstack([np.ones((T, 1)), X_test])
def mul_linear_regresstion(y, x, n , p):
    XtX_inv = np.linalg.inv(x.T @ x)
    B = XtX_inv @ x.T @ y
    return B
def result(x, B):
    y_hat = x @ B
    y_hat = np.round(y_hat, 2)
    return y_hat
B = mul_linear_regresstion(y,X,N,F)
y_hat = np.array(result(X_test, B))
for val in y_hat:
    print(f"{val:}")
```
